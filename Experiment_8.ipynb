Practical 8: Perform various Spatial domain filtering techniques on below-attached image
(filter.tif)
Apply filter operation with a different kernel sizes of 3x3, 7x7, 11x11,15x15 and 21x21
Low pass filtering
High pass filtering
Weighted average filter

```python
from skimage import io
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from matplotlib.colors import LogNorm
from scipy import signal
```

```python
img = cv2.imread('filter.tif', cv2.IMREAD_UNCHANGED)
#img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img = cv2.resize(img, [200,200])
cv2_imshow(img)
```

```python
mask_size = [3, 7, 11, 15, 21]
m, n = img.shape
```

```python
img_new = np.zeros([m, n])
```

```python
#low pass filter
for x in mask_size:
  mask = np.ones([x, x], dtype = int)
  mask = mask / 9
  for i in range(1, m-1):
    for j in range(1, n-1):
        temp = img[i-1, j-1]*mask[0, 0]+img[i-1, j]*mask[0, 1]+img[i-1, j + 1]*mask[0, 2]+img[i, j-1]*mask[1, 0]+ img[i, j]*mask[1, 1]+
img[i, j + 1]*mask[1, 2]+img[i + 1, j-1]*mask[2, 0]+img[i + 1, j]*mask[2, 1]+img[i + 1, j + 1]*mask[2, 2]
        
        img_new[i, j]= temp
         
  img_new = img_new.astype(np.uint8)
  cv2_imshow(img_new)
```

```python
#High Pass Filter
for x in mask_size:
  mask = np.ones([x, x], dtype = int)
  mask = mask * (-1/9)
  mask[int(x/2)][int(x/2)]=8/9
  for i in range(1, m-1):
    for j in range(1, n-1):
        temp = img[i-1, j-1]*mask[0, 0]+img[i-1, j]*mask[0, 1]+img[i-1, j + 1]*mask[0, 2]+img[i, j-1]*mask[1, 0]+ img[i, j]*mask[1, 1]+img[i, j + 1]*mask[1, 2]+img[i + 1, j-1]*mask[2, 0]+img[i + 1, j]*mask[2, 1]+img[i + 1, j + 1]*mask[2, 2]
        
        img_new[i, j]= temp
         
  img_new = img_new.astype(np.uint8)
  cv2_imshow(img_new)
```

```python
#Weighted average filter
for x in mask_size:
  img3=signal.medfilt2d(img, kernel_size=x)
  cv2_imshow(img3)
```
