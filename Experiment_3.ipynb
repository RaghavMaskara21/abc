Take one particular colour image, convert it into a grayscale image and perform the following transformation operations:
     a) Log transform (for c=10 and c=30)
     b) Power law transform (c=1)
         (For dark image , λ=0.6, 0.2)
         (For brighter image, λ=4, 8)
     c) Contrast stretching

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

img = cv2.imread('pic1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.resize(img, [200,200]) 
cv2_imshow(img)
```

```python
#Log transform for c=10 and c=30
c = 10
c_10_img = c * (np.log(img + 1))
cv2_imshow(c_10_img)
c = 30
c_30_img = c * (np.log(img + 1))
cv2_imshow(c_30_img)
```

```python
#Power law transform (c=1) (For dark image , λ=0.6, 0.2) 
def PowTransform(img, gamma, c=1):
  return 255*c*(img/255)**gamma
cv2_imshow(PowTransform(img, gamma=0.6, c=1))
cv2_imshow(PowTransform(img, gamma=0.2, c=1))
```

```python
#(For brighter image, λ=4, 8)
cv2_imshow(PowTransform(img, gamma=4, c=1))
cv2_imshow(PowTransform(img, gamma=8, c=1))
```

```python
#Contrast stretching
def ContrastStretching(img):
    min_percent = 2   # Low percentile
    max_percent = 98  # High percentile
    low, high = np.percentile(img, (min_percent, max_percent))

    return (img-low)/(high-low)*255
cv2_imshow(img)
cv2_imshow(ContrastStretching(img))
```
