Take a colour image & perform the following operation on Gray scale image:
(A)  Brightness Increases or Decreases
(B)  Contrast Increases or Decreases
(C)  Negative Image
(D)  Take two colour images, convert them into grey scale images, resize them and perform addition, subtraction, multiplication and division operations.

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('pic1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.resize(img, [200,200]) 
cv2_imshow(img)
```

```python
#increase decrease brightness
def brightness(img, inc):
  return np.clip(img+inc, 0 ,255)

cv2_imshow(np.clip(brightness(img, 50),0,255))
cv2_imshow(np.clip(brightness(img, -50),0,255))
```

```python
#increase decrease contrast
def contrast(img, cent):
  return img+cent*(img)

cv2_imshow(np.clip(contrast(img, 0.8), 0, 255))
cv2_imshow(np.clip(contrast(img, -0.8), 0, 255))
```

```python
#create negative image
def negative(img):
  return 255-img

cv2_imshow(negative(img))
```

```python
img2 = cv2.imread("img2.jpg")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img2 = cv2.resize(img2, [200,200])
cv2_imshow(img2)
```

```python
def addition(img1, img2):
  return np.clip(img1+img2, 0 ,255)
cv2_imshow(addition(img , img2))
```

```python
def subtraction(img1, img2):
  return np.clip(img1-img2, 0 ,255)
cv2_imshow(subtraction(img , img2))
```

```python
def multiplication(img1, img2):
  return np.clip(img1*img2, 0 ,255)
cv2_imshow(multiplication(img , img2))
```

```python
def division(img1, img2):
  return np.clip(img1/img2, 0 ,255)
cv2_imshow(division(img , img2))
```
