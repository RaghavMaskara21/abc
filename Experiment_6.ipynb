Take Grayscale or binary image and perform the following: Morphological
operations .
1. Erosion

 cv2.erode(img,kernel,iterations=1)
2. Dilation

    cv2.dilate(img,kernel,iterations=1)
3. Opening
     cv2.morphology Ex(img,cv2.MORPH_OPEN,kernel)
4. Closing
     cv2.morphology Ex(img,cv2.MORPH_CLOSE,kernel)

```python
from skimage import io
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from matplotlib.colors import LogNorm
```

```python
img = cv2.imread('pic1.jpg',cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, [200,200]) 
cv2_imshow(img)
```

```python
binary_image = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
kernel = np.ones((3, 3), np.uint8)
inverted_image = cv2.bitwise_not(binary_image)
```

```python
# Erosion
erosion = cv2.erode(binary_image, kernel, iterations=1)
plt.imshow(erosion, cmap='gray')
```

```python
# Dilation
dilation = cv2.dilate(binary_image, kernel, iterations=1)
plt.imshow(dilation, cmap='gray')
```

```python
# opening
opening = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, kernel, iterations=1)
plt.imshow(opening, cmap='gray')
```

```python
# Closing
closing = cv2.morphologyEx(binary_image, cv2.MORPH_CLOSE, kernel, iterations=1)
plt.imshow(closing, cmap='gray')
```
