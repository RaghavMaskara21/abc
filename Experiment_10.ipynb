```python
import cv2
import numpy as np
import cmath
import matplotlib.pyplot as plt
import scipy
import scipy.ndimage
```

```python
img1 = cv2.imread('pic1.jpg',0)
img2 = cv2.imread('pic2.jpg',0)
img2 = cv2.resize(img2,[600,600])
```

```python
f = np.fft.fft2(img1)
fshift1 = np.fft.fftshift(f)
phase_spectrumA = np.angle(fshift1)
magnitude_spectrumA = 20*np.log(np.abs(fshift1))
```

```python
plt.subplot(121),plt.imshow(img1, cmap = 'gray')
plt.title('Input Image A'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrumA, cmap = 'gray')
plt.title('Magnitude Spectrum A'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
f2 = np.fft.fft2(img2)
fshift2 = np.fft.fftshift(f2)
phase_spectrumB = np.angle(fshift2)
magnitude_spectrumB = 20*np.log(np.abs(fshift2))
```

```python
plt.subplot(121),plt.imshow(img2, cmap = 'gray')
plt.title('Input Image B'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(phase_spectrumA, cmap = 'gray')
plt.title('Phase Spectrum B'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
combined1 = np.multiply(magnitude_spectrumA,phase_spectrumB)

plt.subplot(121),plt.imshow(combined1, cmap = 'gray')
plt.title('magnitudeA X phaseB'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
combined2 = np.multiply(phase_spectrumA,magnitude_spectrumB)

plt.subplot(121),plt.imshow(img1, cmap = 'gray')
plt.title('Input Image A'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(phase_spectrumA, cmap = 'gray')
plt.title('phase Spectrum A'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
plt.subplot(121),plt.imshow(img2, cmap = 'gray')
plt.title('Input Image B'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrumA, cmap = 'gray')
plt.title('Magnitude Spectrum B'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
plt.subplot(121),plt.imshow(combined2, cmap = 'gray')
plt.title('phaseA x magnitudeB'), plt.xticks([]), plt.yticks([])
plt.show()
```
