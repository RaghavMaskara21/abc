Take one particular colour image, convert it into a gray-scale image and perform the following transformation operations:
   (a) Gray level slicing (with and without background)
   (b) Bit Plane slicing


```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('pic1.jpg', cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, [200,200])
cv2_imshow(img)
```

```python
#grey level slicing with background 
m,n = img.shape[:2] 
img_thresh_back = np.zeros((m,n), dtype = int) 
T1=100 
T2=180 
for i in range(m): 
  for j in range(n): 
    if T1 < img[i,j] < T2: 
      img_thresh_back[i,j]= 255 
    else: 
      img_thresh_back[i,j] = img[i,j] 
cv2_imshow(img_thresh_back)
```

```python
#grey level slicing without background 
m,n = img.shape[:2] 
img_thresh_back = np.zeros((m,n), dtype = int) 
T1=100 
T2=180 
for i in range(m): 
  for j in range(n): 
    if T1 < img[i,j] < T2: 
      img_thresh_back[i,j]= 255 
    else: 
      img_thresh_back[i,j] = 0
cv2_imshow(img_thresh_back)
```

```python
# Bit Plane slicing
eight_bit_img = img & 2**7
seven_bit_img = img & 2**6
six_bit_img = img & 2**5
five_bit_img = img & 2**4
four_bit_img = img & 2**3
three_bit_img = img & 2**2
two_bit_img = img & 2**1
one_bit_img = img & 2**0
```

```python
#Concatenate these images 
p1 = cv2.hconcat([eight_bit_img,seven_bit_img,six_bit_img,five_bit_img])
p2 =cv2.hconcat([four_bit_img,three_bit_img,two_bit_img,one_bit_img])
output = cv2.vconcat([p1,p2])
cv2_imshow(output)
```
