Take one particular colour image, convert it into a gray-scale image and perform the following transformation operations:
   (a) Gray level slicing (with and without background)
   (b) Bit Plane slicing


```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('pic1.jpg', cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, [200,200])
cv2_imshow(img)
```

```python
def GraySlicing(temp, low, high, background=True):
  if not background:
    temp1=temp
    temp[temp1>high]=255
    temp[temp1<low]=255
    return temp
  else:
    temp[(temp>high) & (temp<low)]=255
    return temp
```

```python
#Gray level slicing (with and without background)
without_background = GraySlicing(temp=img, low=30, high= 200, background=True)
with_background = GraySlicing(temp=img, low=30, high=200, background=False)
res = np.hstack([with_background, without_background])
cv2_imshow(res)
```

```python
# Bit Plane slicing
eight_bit_img = img & 2**7
seven_bit_img = img & 2**6
six_bit_img = img & 2**5
five_bit_img = img & 2**4
four_bit_img = img & 2**3
three_bit_img = img & 2**2
two_bit_img = img & 2**1
one_bit_img = img & 2**0
```

```python
#Concatenate these images 
p1 = cv2.hconcat([eight_bit_img,seven_bit_img,six_bit_img,five_bit_img])
p2 =cv2.hconcat([four_bit_img,three_bit_img,two_bit_img,one_bit_img])
output = cv2.vconcat([p1,p2])
cv2_imshow(output)
```
