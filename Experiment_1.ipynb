(A)   Take a colour image and perform following conversions:

(i)  RGB to Grey (ii) RGB to index and (iii) RGB to Binary (with and without predefined function)

(B)   Construct Black image, White image, Black in White, White in Black image.


```python
from PIL import Image
import cv2
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img  = cv2.imread("pic1.jpg", cv2.IMREAD_UNCHANGED)
img = cv2.resize(img, [200,200]) 
```

```python
#Original image
cv2_imshow(img)
```

```python
#RGB to Grey
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

```python
#RGB to index
index_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2_imshow(index_img)
```

```python
#RGB to Binary with predefined function
ret, binary_img_fn = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)
cv2_imshow(binary_img_fn)
```

```python
#RGB to Binary without predefined function
import numpy as np
def ConvertGrayToBinary(img, thresh):
    return 255*(np.array(img)>thresh)

cv2_imshow(ConvertGrayToBinary(gray_img, 127))
```

```python
#Black image
black_image = np.zeros([200, 200])
cv2_imshow(black_image)
```

```python
#white image
white_image = 255*np.ones([200, 200])  
cv2_imshow(white_image)
```

```python
#Black in White
bw = white_image
bw[50:150, 50:150] = 0
cv2_imshow(bw)
```

```python
#White in Black
wb = black_image
wb[50:150, 50:150] = 255
cv2_imshow(wb)
```

```python
#To plot 2D function f(x, y) = a sin(u(x)*x + v(y)*y), where u and v are random
import math
import random
import numpy as np
from google.colab.patches import cv2_imshow

img = np.zeros([256,256])
u =  np.random.rand(256)
v =  np.random.rand(256)

def function(x, y, u, v):
  a = 255  
  return a*math.sin((u[x]*x) + (v[y]*y))

for i in range(256):
  for j in range(256):
    img[i,j] = function(i, j, u, v)

cv2_imshow(img)
```
